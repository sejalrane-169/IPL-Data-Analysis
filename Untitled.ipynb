```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
print("Libraries loaded successfully!")
```

```python
# Load all 4 files
matches = pd.read_csv("matches.csv")
deliveries = pd.read_csv("deliveries.csv")
players = pd.read_csv("players.csv")
seasons = pd.read_csv("seasons.csv")

# Fix missing values
matches['winner'] = matches['winner'].fillna('No Result')
matches['player_of_match'] = matches['player_of_match'].fillna('Unknown')

# Fix team names that changed over the years
team_name_fix = {
    'Delhi Daredevils': 'Delhi Capitals',
    'Kings XI Punjab': 'Punjab Kings',
    'Rising Pune Supergiant': 'Rising Pune Supergiants',
    'Deccan Chargers': 'Sunrisers Hyderabad'
}
matches['team1'] = matches['team1'].replace(team_name_fix)
matches['team2'] = matches['team2'].replace(team_name_fix)
matches['winner'] = matches['winner'].replace(team_name_fix)
deliveries['batting_team'] = deliveries['batting_team'].replace(team_name_fix)
deliveries['bowling_team'] = deliveries['bowling_team'].replace(team_name_fix)

print("Data cleaned successfully!")
print(f"Total matches: {len(matches)}")
print(f"Total deliveries: {len(deliveries)}")
```

```python
# Count wins per team, exclude 'No Result'
win_counts = matches[matches['winner'] != 'No Result']['winner'].value_counts().head(10)

# Draw the chart
plt.figure(figsize=(12, 6))
# figsize=(12,6) means 12 inches wide, 6 inches tall

win_counts.plot(kind='bar', color='steelblue', edgecolor='black')

plt.title('Top 10 Teams by Total Wins in IPL History', fontsize=14)
plt.xlabel('Team')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45, ha='right')
# rotation=45 tilts team names so they don't overlap
plt.tight_layout()
# tight_layout stops labels from getting cut off

plt.savefig('top_teams_wins.png', dpi=150)
# saves chart as image file 
plt.show()
```

```python
# Create a new column — True if toss winner also won the match
matches['toss_helped'] = matches['toss_winner'] == matches['winner']

# Count True vs False
toss_result = matches['toss_helped'].value_counts()

plt.figure(figsize=(6, 6))
plt.pie(
    toss_result,
    labels=['Toss Winner Won', 'Toss Winner Lost'],
    autopct='%1.1f%%',
    # autopct shows percentage on each slice
    colors=['#2ecc71', '#e74c3c'],
    startangle=90
)
plt.title('Does Winning the Toss Help Win the Match?', fontsize=14)
plt.savefig('toss_impact.png', dpi=150)
plt.show()
```

```python
toss_decision = matches['toss_decision'].value_counts()

plt.figure(figsize=(6, 5))
toss_decision.plot(
    kind='bar',
    color=['#3498db', '#e67e22'],
    edgecolor='black'
)
plt.title('What Do Teams Choose After Winning the Toss?', fontsize=14)
plt.ylabel('Number of Times')
plt.xticks(rotation=0)
plt.tight_layout()
plt.savefig('toss_decision.png', dpi=150)
plt.show()
```

```python
venue_counts = matches['venue'].value_counts().head(10)

plt.figure(figsize=(12, 6))
venue_counts.plot(kind='barh', color='coral', edgecolor='black')
# barh = horizontal bar chart — better for long venue names

plt.title('Top 10 Venues by Number of Matches Hosted', fontsize=14)
plt.xlabel('Number of Matches')
plt.tight_layout()
plt.savefig('top_venues.png', dpi=150)
plt.show()
```

```python
# Group all deliveries by batsman, sum their runs
top_batsmen = deliveries.groupby('striker')['batsman_runs'].sum()
top_batsmen = top_batsmen.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_batsmen.plot(kind='bar', color='gold', edgecolor='black')
plt.title('Top 10 Batsmen by Total IPL Runs (2008-2025)', fontsize=14)
plt.xlabel('Player')
plt.ylabel('Total Runs')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_batsmen.png', dpi=150)
plt.show()
```

```python
# Filter only actual wickets, exclude run outs (those aren't bowler's credit)
wickets = deliveries[
    (deliveries['is_wicket'] == 1) &
    (deliveries['dismissal_type'] != 'run out')
]

top_bowlers = wickets.groupby('bowler')['is_wicket'].count()
top_bowlers = top_bowlers.sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
top_bowlers.plot(kind='bar', color='tomato', edgecolor='black')
plt.title('Top 10 Bowlers by Total IPL Wickets (2008-2025)', fontsize=14)
plt.xlabel('Player')
plt.ylabel('Total Wickets')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('top_bowlers.png', dpi=150)
plt.show()
```

```python
# This uses your seasons.csv file
plt.figure(figsize=(14, 6))
champion_counts = seasons['champion'].value_counts()

champion_counts.plot(kind='bar', color='purple', edgecolor='black')
plt.title('IPL Championships Won by Each Team', fontsize=14)
plt.xlabel('Team')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('ipl_champions.png', dpi=150)
plt.show()
```

```python
print("=" * 50)
print("KEY INSIGHTS FROM IPL DATA 2008-2025")
print("=" * 50)

# Most wins
top_team = matches[matches['winner'] != 'No Result']['winner'].value_counts().index[0]
top_wins = matches[matches['winner'] != 'No Result']['winner'].value_counts().values[0]
print(f"\n1. {top_team} has the most wins in IPL history with {top_wins} victories")

# Toss impact
toss_helped_pct = round(matches['toss_helped'].mean() * 100, 1)
print(f"\n2. Toss winners went on to win the match {toss_helped_pct}% of the time")

# Most popular toss decision
top_decision = matches['toss_decision'].value_counts().index[0]
top_decision_pct = round(matches['toss_decision'].value_counts().values[0] / len(matches) * 100, 1)
print(f"\n3. {top_decision_pct}% of toss winners chose to {top_decision} first")

# Top batsman
top_bat = deliveries.groupby('striker')['batsman_runs'].sum().sort_values(ascending=False)
print(f"\n4. {top_bat.index[0]} is the all-time leading run scorer with {top_bat.values[0]:,} runs")

# Top bowler
print(f"\n5. {top_bowlers.index[0]} leads all bowlers with {top_bowlers.values[0]} wickets")

print("\n" + "=" * 50)
```
